```python
import matplotlib.pyplot as plt
x = ['Fict', 'Tech', 'Moti', 'Business', 'Nutri', 'Dev']
sales = [5.2,19.6, 8.7, 8, 7.7, 3.7]
x_pos = [i for i, _ in enumerate(x)]
#label = ['Fict', 'Tech', 'Moti', 'Business', 'Nutri', 'Dev']
plt.bar(x[0], sales[0], color='red', label = 'Fict')
plt.bar(x[1], sales[1], color='black', label = 'Tech')
plt.bar(x[2], sales[2], color='cyan', label = 'Moti')
plt.bar(x[3], sales[3], color='green', label = 'Business')
plt.bar(x[4], sales[4], color='yellow', label = 'Nutri')
plt.bar(x[5], sales[5], color='pink', label = 'Dev')
plt.xlabel("Books")
plt.ylabel("Sales")
plt.title("Sales of book")
plt.xticks(x_pos, x)
plt.legend(loc='upper right')
# Turn on the grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = (22, 30, 35, 35, 26)
womenMeans = (25, 32, 30, 35, 29)
menStd = (4, 3, 4, 1, 5)
womenStd = (3, 5, 2, 3, 3)
# the x locations for the groups
ind = np.arange(N)    
# the width of the bars
width = 0.35      

p1 = plt.bar(ind, womenMeans, width, yerr=menStd, color='red')
p2 = plt.bar(ind, menMeans, width,
bottom=womenMeans, yerr=womenStd, color='green')

plt.ylabel('Scores')
plt.xlabel('Groups')
plt.title('Scores by group\n' + 'and gender')
plt.xticks(ind, ('Group1', 'Group2', 'Group3', 'Group4', 'Group5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Women', 'Men'))
plt.annotate("Annotation", (1,40), xycoords='data',xytext=(1,70),arrowprops=dict(arrowstyle='->'))

plt.show()
```
