```python
import seaborn as sns
from matplotlib.pyplot import *
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
sns.displot(data=titanic,x="age",hue="alive",multiple="stack")
```

```python
sns.displot(data=titanic,x="survived",hue="pclass",kind="kde")
```

```python
sns.kdeplot(data=titanic,x="survived",hue="pclass")
```

```python
sns.lineplot(data=titanic,x="survived",y="pclass",lw=2,hue="who",style="alone",size="who")
```

```python
sns.scatterplot(data=titanic,x="fare",y="age",hue="pclass",size="pclass",sizes=(50,200),palette="deep")
```

```python
sns.scatterplot(data=titanic,x="fare",y="age",hue="pclass",size="pclass",sizes=(50,200),palette="flare")
```

```python
sns.scatterplot(data=titanic,x="fare",y="age",hue="pclass",size="pclass",sizes=(50,200),palette="pastel")
```
