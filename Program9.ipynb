```python
import seaborn as sns
from matplotlib.pyplot import *
```

```python
titan=sns.load_dataset("titanic")
```

```python
titan.head()
```

```python
sns.relplot(data=titan,x='class', y='fare', kind='scatter', hue='embark_town', row='sex')
```

```python
sns.catplot(data=titan,x='class', y='fare', kind='violin', hue='embark_town')
```

```python
sns.lineplot(data=titan, x='survived', y='class', hue='sex' )
```

```python
sns.catplot(data=titan, x='alive', y='fare', kind='strip', jitter=False )
```
