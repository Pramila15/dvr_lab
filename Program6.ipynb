```python
import seaborn as sns
import numpy as np
```

```python
planet=sns.load_dataset("planets")
```

```python
planet
```

```python
sns.relplot(data=planet,x="distance",y="year",kind="scatter")
```

```python
sns.scatterplot(data=planet,x="distance",y="year",)
```

```python
sns.relplot(data=planet,x="distance",y="year",kind="scatter",hue="mass")
```

```python
sns.relplot(data=planet,x="distance",y="year",kind="scatter",hue="mass",size="mass")
```

```python
sns.relplot(data=planet,x="distance",y="year",kind="scatter",hue="mass",marker=True)
```

```python
sns.relplot(data=planet,x="distance",y="year",kind="scatter",marker="+",size="mass")
```

```python
sns.relplot(data=planet,x="distance",y="year",kind="scatter",markers=True,s=100)
```

```python
sns.relplot(data=planet,x="distance",y="year",kind="scatter",markers=True,size="mass",sizes=(20,100))
```

```python
sns.displot(x="orbital_period",row="year",data=planet,kind="hist",bins=50)
```
