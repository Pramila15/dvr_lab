```python
import seaborn as sns
from matplotlib.pyplot import *
```

```python
titan=sns.load_dataset("titanic")
```

```python
titan.head()
```

```python
sns.catplot(x="pclass", data=titan, kind="count", col="embarked")
sns.catplot(x="pclass", data=titan, kind="count", col="embarked", col_order=['S', 'Q', 'C'])
```

```python
sns.catplot(data=titan,x='class', y='fare', kind='violin', hue='embark_town')
```

```python
sns.relplot(data=titan,x='pclass', y='alive', kind='line', hue='embark_town', col='pclass')
```

```python
sns.relplot(data=titan, x='class', y='fare', kind='scatter', hue='embark_town', palette='bright', s=100)
```
