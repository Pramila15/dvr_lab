```python
import seaborn as sns
import numpy as np
```

```python
planet=sns.load_dataset("planets")
```

```python
planet
```

```python
sns.relplot(data=planet, x='distance', y='year', kind='scatter', hue='mass', size='method')
```

```python
sns.relplot(data=planet, x='distance', y='year', kind='scatter', hue='method', style='method', s=100)
```

```python
sns.barplot(data=planet, x='distance', y='method', hue = 'method')
```
